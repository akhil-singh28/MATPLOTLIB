**2. SCATTER PLOT**

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5,6,7]
Num=[3,6,1,2,4,7,5]
plt.scatter(Day,Num,color="r")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("Nums",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5,6,7]
Num=[3,6,1,2,4,7,5]
colors=["r","y","g","b","m","cyan","black"]
size=[100,20,44,50,10,35,70]
plt.scatter(Day,Num,c=colors,s=size,alpha=0.5)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("Nums",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5,6,7]
Num=[3,6,1,2,4,7,5]
colors=["r","y","g","b","m","cyan","pink"]
size=[100,200,440,500,100,350,650]
plt.scatter(Day,Num,c=colors,s=size,marker="*",edgecolor="black",linewidth=2)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("Nums",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5,6,7]
Num=[3,6,1,2,4,7,5]
colors=[10,49,30,29,56,20,30]
size=[100,200,440,500,100,350,650]
plt.scatter(Day,Num,c=colors,s=size,cmap="CMRmap")
t=plt.colorbar()
t.set_label("Color Code",fontsize=15)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("Nums",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
Day=[1,2,3,4,5,6,7]
Num=[3,6,1,2,4,7,5]
Num2=[3,2,4,5,1,6,2]
colors=[10,49,30,29,56,20,30]
size=[100,200,440,500,100,350,650]
plt.scatter(Day,Num,c=colors,s=size,cmap="CMRmap")
plt.scatter(Day,Num2,color="r",s=size)
t=plt.colorbar()
t.set_label("Color Code",fontsize=15)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=10)
plt.ylabel("Nums",fontsize=10)
plt.show()
```
