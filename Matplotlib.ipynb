**BASIC FIRST PLOT**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
c=["r",'y',"g",'b']
plt.bar(x,y,color=c)
plt.show()
```

**1.MATPLOTLIB BAR PLOT**

```python
import matplotlib.pyplot as plt
x=["Python","Java","Rust","JavaScript"]
y=[90,80,60,83]
z=[25,35,20,60]
plt.xlabel("Language",fontsize=10)
plt.ylabel("No.",fontsize=15)
plt.title("PROGRAMMING LANGUAGES",fontsize=10)
c=["magenta","blue","red","cyan"]
# plt.bar(x,y,width=0.4,color=c,align="center",edgecolor="black",linewidth=2,linestyle=":",alpha=0.50)  

plt.bar(x,y,width=0.4,color=c,edgecolor="black",label="Language")
plt.bar(x,z,width=0.4,color="yellow",label="Popularity")

#alpha value ranges between (0 to 1)
#align="edge"

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["Python","Java","Rust","JavaScript"]
y=[90,80,60,83]
z=[25,35,20,60]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language",fontsize=10)
plt.ylabel("No.",fontsize=15)
plt.title("PROGRAMMING LANGUAGES",fontsize=10)
# c=["magenta","blue","red","cyan"]

plt.bar(p,y,width,color=c,edgecolor="black",label="Language")
plt.bar(p1,z,width,color="yellow",edgecolor="green",label="Popularity")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
