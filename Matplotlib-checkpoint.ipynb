**MATPLOTLIB**

Matplotlib is one of the most widely used Python libraries for data visualization. It lets you create line plots, bar charts, scatter plots, histograms, pie charts, heatmaps, and much more.

**BASIC LINE PLOT**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
